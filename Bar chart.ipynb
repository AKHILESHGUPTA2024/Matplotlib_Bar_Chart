#### Bar chart
###### Bar chart mainly deals with categorical fields

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#simple bar chart
#create data
fig=plt.figure(figsize=(12,4))
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales =  [29,49,83,58,45,76,94,20,72,64,50,56]
x=plt.bar(months,sales,color='grey')
plt.bar_label(x,sales,fontsize=6)
plt.title('Sales Discipline')
plt.xlabel('months',fontsize=6)
plt.ylabel('sales',fontsize=6)
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)
```

```python
#side by side bar chart
fig=plt.figure(figsize=(12,4))
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales =  [29,49,83,58,45,76,94,20,72,64,50,56]
sales1 = [20,39,43,18,35,56,56,8,42,44,35,36]
barwidth=0.4
m=np.arange(len(months))
x=plt.bar(m,sales,width=barwidth,color='grey')
y=plt.bar(m+barwidth,sales1,width=barwidth,color='blue')
plt.title('sales over months',fontsize=12)
plt.xlabel('months')
plt.ylabel('sales')
plt.bar_label(x,sales,fontsize=6)
plt.bar_label(y,sales1,fontsize=6)
plt.grid(alpha=0.5)
plt.xticks(m+0.2,months,fontsize=6)
plt.yticks(fontsize=6)
```

```python
#stacked bar chart
fig=plt.figure(figsize=(12,4))
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales =  [29,49,83,58,45,76,94,20,72,64,50,56]
sales1 = [20,39,43,18,35,56,56,8,42,44,35,36]
barwidth=0.4
m=np.arange(len(months))
x=plt.bar(m,sales,width=barwidth,color='grey')
y=plt.bar(m,sales1,width=barwidth,color='blue')
plt.title('sales over months',fontsize=12)
plt.xlabel('months')
plt.ylabel('sales')
plt.bar_label(x,sales,fontsize=6)
plt.bar_label(y,sales1,fontsize=6)
plt.grid(alpha=0.5)
plt.xticks(m+0.2,months,fontsize=6)
plt.yticks(fontsize=6)
```
